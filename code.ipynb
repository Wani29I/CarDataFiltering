```python
def readFile(filePath):
  df = pd.read_excel(filePath, skiprows=1)
  return df

def checkTime(eachRow):
    errorTime = False
    
    if(eachRow["End time - OT"].hour not in timeKeys):
        errorTime = True

    time = eachRow["End time - OT"].hour

    return time, errorTime

def checkRoute(eachRow):
    errorRoute = True
    route = eachRow['Transportation route'].strip()
    for route_name, keys in allRouteKeys.items():
        if route in keys:
            route = route_name
            errorRoute = False
            break
    return route, errorRoute

def createFinalDataDict():
    times = timeKeys
    routes = range(1, len(routeName)+1)  # Creates numbers 1 to 14

    # Generate the dictionary automatically
    FinalDataDict = {}
    for r in routes:
        FinalDataDict[r] = {}
        for t in times:
            FinalDataDict[r][t] = []

    return FinalDataDict

def mainFilter(df):
    FinalDataDict = createFinalDataDict()
    errorTimeList = []
    errorRouteList = []

    # loop to every data
    for index, eachRow in df.iterrows():
        # check and get time
        time, errorTime = checkTime(eachRow)
        # check and get route
        route, errorRoute = checkRoute(eachRow)

        # if not error in time and route: add to filtered one
        if(not errorTime and not errorRoute):
            FinalDataDict[route][time].append(index)

        # if error append to error list
        if(errorTime):
            errorTimeList.append([index+3, eachRow["End time - OT"]])
        if(errorRoute):
            errorRouteList.append([index+3, eachRow['Transportation route'].strip()])

    return FinalDataDict, errorTimeList, errorRouteList

def createFinalExcelList(FinalDataDict, df):
    countAll = 0
    finalIndexList = []
    # create blank row (3 rows) prepare for append
    blankRow3 = pd.DataFrame([[np.nan] * len(df.columns)]*3, columns=df.columns)

    # loop into each time 8,17,20
    for route in FinalDataDict:
        # loop into each route
        for time in FinalDataDict[route]:
            # count
            countAll += len(FinalDataDict[route][time])
            # get df for only current data
            currentData = df.iloc[FinalDataDict[route][time]]
            # add ID column for current data
            currentData.insert(1, 'ID', range(1, len(currentData) + 1))
            # strift time
            # currentData['End time - OT'] = currentData['End time - OT'].dt.strftime('%H:%M')

            # Create new summary data row
            header_data = [f"Route: {routeName[route]}", f"Time Slot: {time}", f"Count: {len(FinalDataDict[route][time])}"] + [None] * (len(currentData.columns) - 3)
            header_df = pd.DataFrame([header_data], columns=currentData.columns)
            # 2. Append the header DataFrame to your list
            finalIndexList.append(header_df)
            # append to final list
            finalIndexList.append(currentData)
            # append blank row
            finalIndexList.append(blankRow3)
    return finalIndexList, countAll

def checkError(countAll, errorTimeList, errorRouteList):
    print(f"Filtered Data Count: {countAll}")

    if(len(errorTimeList) == 0):
        print("\nNo Time Error <3")
    else:
        print(f"\nTime Error: {len(errorTimeList)}")
        for timeError in errorTimeList:
            print(f"Index: {timeError[0]}  ----------  Time: {timeError[1]}")

    if(len(errorRouteList) == 0):
        print("\nNo Route Error <3")
    else:
        print(f"\nRoute Error: {len(errorRouteList)}")
        for routeError in errorRouteList:
            print(f"Index: {routeError[0]}  ----------  Route Name: {routeError[1]}")

def saveOutputFile(finalIndexList, outputFilename = 'FinalData.xlsx'):
    finalDf = pd.concat(finalIndexList, ignore_index=True)

    if os.path.exists(outputFilename):
        try:
            os.remove(outputFilename)
            print("Removed", outputFilename)
        except PermissionError:
            print(f"Error: Cannot overwrite '{outputFilename}'. Please close the file in Excel first!")
            exit()

    finalDf.to_excel(outputFilename, index=False)
    print(f"Successfully overwritten {outputFilename}")
```

```python
df = readFile('./Book1.xlsx')
FinalDataDict, errorTimeList, errorRouteList = mainFilter(df)
finalIndexList, countAll = createFinalExcelList(FinalDataDict, df)
```

```python
checkError(countAll, errorTimeList, errorRouteList)
```

```python
saveOutputFile(finalIndexList, outputFilename = 'FinalData.xlsx')
```
